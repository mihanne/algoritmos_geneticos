```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt

class Individual:
    def __init__(self, n_variables, lower_bound, upper_bound):
        self.n_variables = n_variables
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.chromosome = self._generate_chromosome()
        self.fitness = 0.0

    def _generate_chromosome(self):
        return [random.uniform(self.lower_bound, self.upper_bound) for _ in range(self.n_variables)]

    def calculate_fitness(self, objective_function, problem_type):
        self.fitness = objective_function(self.chromosome)
        if problem_type == "min":
            if self.fitness >= 0:
                self.fitness = 1 / (1 + self.fitness)

    def __repr__(self):
        return f"Chromosome: {self.chromosome}, Fitness: {self.fitness:.4f}"

class GeneticAlgorithm:
    def __init__(self, population_size, generations, crossover_rate, mutation_rate,
                 n_variables, lower_bound, upper_bound, objective_function, problem_type):
        self.population_size = population_size
        self.generations = generations
        self.crossover_rate = crossover_rate
        self.mutation_rate = mutation_rate
        self.n_variables = n_variables
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.objective_function = objective_function
        self.problem_type = problem_type
        self.population = []
        self.best_individual_overall = None
        self.best_fitness_overall = float('-inf') if problem_type == "max" else float('inf')

    def _initialize_population(self):
        self.population = [Individual(self.n_variables, self.lower_bound, self.upper_bound) for _ in range(self.population_size)]
        self._evaluate_population()

    def _evaluate_population(self):
        for individual in self.population:
            individual.fitness = self.objective_function(individual.chromosome)

    def _select_parents(self):
        fitness_values = [ind.fitness for ind in self.population]
        if self.problem_type == "min":
             min_val = min(fitness_values)
            if min_val < 0:
                adjusted_fitness = [f - min_val + 1e-6 for f in fitness_values]
            else:
                max_val = max(fitness_values)
                adjusted_fitness = [max_val - f + 1e-6 for f in fitness_values]
        else:
            adjusted_fitness = [f + 1e-6 for f in fitness_values]
        total_fitness = sum(adjusted_fitness)
        if total_fitness == 0:
            return random.sample(self.population, 2)
        probabilities = [f / total_fitness for f in adjusted_fitness]
        parents = random.choices(self.population, weights=probabilities, k=2)
        return parents[0], parents[1]

    def _crossover(self, parent1, parent2):
        child1_chromosome = list(parent1.chromosome)
        child2_chromosome = list(parent2.chromosome)
        if random.random() < self.crossover_rate:
            crossover_point = random.randint(1, self.n_variables - 1)
            alpha = random.random()
            for i in range(crossover_point, self.n_variables):
                child1_chromosome[i] = alpha * parent1.chromosome[i] + (1 - alpha) * parent2.chromosome[i]
                child2_chromosome[i] = alpha * parent2.chromosome[i] + (1 - alpha) * parent1.chromosome[i]
            child1_chromosome = [max(self.lower_bound, min(self.upper_bound, val)) for val in child1_chromosome]
            child2_chromosome = [max(self.lower_bound, min(self.upper_bound, val)) for val in child2_chromosome]
        child1 = Individual(self.n_variables, self.lower_bound, self.upper_bound)
        child1.chromosome = child1_chromosome
        child2 = Individual(self.n_variables, self.lower_bound, self.upper_bound)
        child2.chromosome = child2_chromosome
        return child1, child2

    def _mutate(self, individual):
        if random.random() < self.mutation_rate:
            mutation_point = random.randint(0, self.n_variables - 1)
            individual.chromosome[mutation_point] = random.uniform(self.lower_bound, self.upper_bound)

    def run(self):
        self._initialize_population()
        history = []
        for generation in range(self.generations):
            new_population = []

            current_best = max(self.population, key=lambda ind: ind.fitness) if self.problem_type == "max" else min(self.population, key=lambda ind: ind.fitness)


            if self.best_individual_overall is None or \
               (self.problem_type == "max" and current_best.fitness > self.best_fitness_overall) or \
               (self.problem_type == "min" and current_best.fitness < self.best_fitness_overall):

                self.best_individual_overall = Individual(self.n_variables, self.lower_bound, self.upper_bound)
                self.best_individual_overall.chromosome = list(current_best.chromosome)
                self.best_individual_overall.fitness = current_best.fitness
                self.best_fitness_overall = current_best.fitness

            new_population.append(current_best)

            while len(new_population) < self.population_size:
                parent1, parent2 = self._select_parents()
                child1, child2 = self._crossover(parent1, parent2)
                self._mutate(child1)
                self._mutate(child2)
                new_population.append(child1)
                if len(new_population) < self.population_size:
                    new_population.append(child2)
            self.population = new_population
            self._evaluate_population()

            best_fitness_in_gen = max(self.population, key=lambda ind: ind.fitness).fitness if self.problem_type == "max" else min(self.population, key=lambda ind: ind.fitness).fitness
            history.append(best_fitness_in_gen)
        return self.best_individual_overall, history

class DifferentialEvolution:
    def __init__(self, D, NP, CR, F, generations, lower_bound, upper_bound, objective_function, problem_type):
        self.D = D  # Number of variables
        self.NP = NP  # Population size
        self.CR = CR  # Crossover rate
        self.F = F  # Differential weight
        self.generations = generations
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.objective_function = objective_function
        self.problem_type = problem_type
        self.population = []
        self.best_individual_overall = None
        self.best_fitness_overall = float('-inf') if problem_type == "max" else float('inf')

    def _initialize_population(self):
        self.population = [Individual(self.D, self.lower_bound, self.upper_bound) for _ in range(self.NP)]
        self._evaluate_population()

    def _evaluate_population(self, individuals=None):
        if individuals is None:
            individuals = self.population
        for individual in individuals:
            individual.fitness = self.objective_function(individual.chromosome)

    def _mutate_and_crossover(self, target_idx):
        target_individual = self.population[target_idx]
        indices = list(range(self.NP))
        indices.pop(target_idx)

        if len(indices) < 3:
            return target_individual.chromosome

        a_idx, b_idx, c_idx = random.sample(indices, 3)
        a = self.population[a_idx].chromosome
        b = self.population[b_idx].chromosome
        c = self.population[c_idx].chromosome

        # Mutation: donor vector v
        donor_vector = [a[j] + self.F * (b[j] - c[j]) for j in range(self.D)]

        # Crossover: binomial crossover
        trial_vector = list(target_individual.chromosome)
        j_rand = random.randint(0, self.D - 1) # Ensure at least one element is from donor_vector

        for j in range(self.D):
            if random.random() < self.CR or j == j_rand:
                trial_vector[j] = donor_vector[j]

            trial_vector[j] = max(self.lower_bound, min(self.upper_bound, trial_vector[j]))
        return trial_vector

    def run(self):
        self._initialize_population()
        history = []

        for generation in range(self.generations):
            new_population = []
            for i in range(self.NP):
                target_individual = self.population[i]
                trial_chromosome = self._mutate_and_crossover(i)

                trial_individual = Individual(self.D, self.lower_bound, self.upper_bound)
                trial_individual.chromosome = trial_chromosome
                self._evaluate_population(individuals=[trial_individual]) # Evaluate only the trial individual


                if self.problem_type == "min":
                    if trial_individual.fitness < target_individual.fitness:
                        new_population.append(trial_individual)
                    else:
                        new_population.append(target_individual)
                else: # problem_type == "max"
                    if trial_individual.fitness > target_individual.fitness:
                        new_population.append(trial_individual)
                    else:
                        new_population.append(target_individual)

            self.population = new_population
            # Update overall best individual after the entire population for the next generation is formed
            current_best = max(self.population, key=lambda ind: ind.fitness) if self.problem_type == "max" else min(self.population, key=lambda ind: ind.fitness)
            if self.best_individual_overall is None or \
               (self.problem_type == "max" and current_best.fitness > self.best_fitness_overall) or \
               (self.problem_type == "min" and current_best.fitness < self.best_fitness_overall):
                self.best_individual_overall = Individual(self.D, self.lower_bound, self.upper_bound)
                self.best_individual_overall.chromosome = list(current_best.chromosome)
                self.best_individual_overall.fitness = current_best.fitness
                self.best_fitness_overall = current_best.fitness

            # Record the best fitness of the current generation for history
            best_fitness_in_gen = max(self.population, key=lambda ind: ind.fitness).fitness if self.problem_type == "max" else min(self.population, key=lambda ind: ind.fitness).fitness
            history.append(best_fitness_in_gen)

        return self.best_individual_overall, history

def quadratic_function(x):
    return sum(val**2 for val in x)

def rastrigin_function(x, A=10):
    n = len(x)
    return A * n + sum(xi**2 - A * math.cos(2 * math.pi * xi) for xi in x)

def run_experiment(algorithm_class, objective_function, problem_type, n_variables, lower_bound, upper_bound,
                   population_size=50, generations=100, crossover_rate=0.8, mutation_rate=0.1,
                   F=0.8): # F for Differential Evolution

    results = []
    histories = []
    for i in range(30):
        if algorithm_class == GeneticAlgorithm:
            algo = algorithm_class(population_size, generations, crossover_rate, mutation_rate,
                                   n_variables, lower_bound, upper_bound, objective_function, problem_type)
        elif algorithm_class == DifferentialEvolution:
            algo = algorithm_class(n_variables, population_size, crossover_rate, F, generations,
                                   lower_bound, upper_bound, objective_function, problem_type)

        best_solution, history = algo.run()
        results.append(best_solution.fitness)
        histories.append(history)

        # print progress for each run
        print(f"Execucao {i+1:02d} - Melhor: {best_solution.fitness:.6f}")

    best_run_result = min(results) if problem_type == "min" else max(results)
    worst_run_result = max(results) if problem_type == "min" else min(results)
    average_result = np.mean(results)
    std_dev_result = np.std(results)

    histories = np.array(histories)
    mean_per_gen = np.mean(histories, axis=0)
    std_per_gen = np.std(histories, axis=0)
    best_per_gen = np.min(histories, axis=0) if problem_type == "min" else np.max(histories, axis=0)
    worst_per_gen = np.max(histories, axis=0) if problem_type == "min" else np.min(histories, axis=0)

    return best_run_result, worst_run_result, average_result, std_dev_result, mean_per_gen, std_per_gen, best_per_gen, worst_per_gen, results

def plot_convergence(mean_gen, std_gen, best_gen, worst_gen, title):
    generations = list(range(1, len(mean_gen) + 1))
    plt.figure(figsize=(10, 6))
    plt.plot(generations, mean_gen, label='Media', color='blue')
    plt.plot(generations, std_gen, label='Desvio padrao', color='gray', linestyle='--') # Std dev can be hard to visualize directly on fitness plot
    plt.plot(generations, best_gen, label='Melhor (na geracao)', color='green')
    plt.plot(generations, worst_gen, label='Pior (na geracao)', color='red')
    plt.title(title)
    plt.xlabel('Geracao')
    plt.ylabel('Fitness')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def count_optimal_hits(results, optimal_value, problem_type, tolerance=1e-4):
    hits = 0
    for res in results:
        if problem_type == "min":
            if abs(res - optimal_value) <= tolerance:
                hits += 1
        else: # problem_type == "max"
            # For maximization, if the optimal value is positive, we check if within tolerance below it
            # If optimal is 0 (or negative), check within tolerance above it.
            # For x^2 with bounds [-10, 10], max is 100. So we check if res is >= (optimal - tolerance)
            if abs(res - optimal_value) <= tolerance:
                 hits += 1
    return hits

if __name__ == "__main__":
    n = 2
    population_size = 50
    generations = 100
    crossover_rate_ga = 0.8
    mutation_rate_ga = 0.1
    crossover_rate_de = 0.8 # CR for DE
    F_de = 0.8 # F for DE
    tolerance = 1e-3

    ga_optimal_hits = {'quadratic_min': 0, 'quadratic_max': 0, 'rastrigin_min': 0}
    de_optimal_hits = {'quadratic_min': 0, 'quadratic_max': 0, 'rastrigin_min': 0}


    # --- Teste 1: Min f1(x) = x^2 ---
    print("\n--- Teste 1 (Algoritmo Genetico): Min f(x) = x^2 ---")
    best_ga, worst_ga, avg_ga, std_ga, mean_gen_ga, std_gen_ga, best_gen_ga, worst_gen_ga, ga_results_quad_min = run_experiment(
        algorithm_class=GeneticAlgorithm,
        objective_function=quadratic_function,
        problem_type="min",
        n_variables=n,
        lower_bound=-10,
        upper_bound=10,
        population_size=population_size,
        generations=generations,
        crossover_rate=crossover_rate_ga,
        mutation_rate=mutation_rate_ga
    )
    print(f"GA - Melhor: {best_ga:.6f}, Pior: {worst_ga:.6f}, Media: {avg_ga:.6f}, Desvio Padrao: {std_ga:.6f}")
    plot_convergence(mean_gen_ga, std_gen_ga, best_gen_ga, worst_gen_ga, "Convergencia GA - Min f(x) = x^2")
    ga_optimal_hits['quadratic_min'] = count_optimal_hits(ga_results_quad_min, 0.0, "min", tolerance)

    print("\n--- Teste 1 (Evolucao Diferencial): Min f(x) = x^2 ---")
    best_de, worst_de, avg_de, std_de, mean_gen_de, std_gen_de, best_gen_de, worst_gen_de, de_results_quad_min = run_experiment(
        algorithm_class=DifferentialEvolution,
        objective_function=quadratic_function,
        problem_type="min",
        n_variables=n,
        lower_bound=-10,
        upper_bound=10,
        population_size=population_size,
        generations=generations,
        crossover_rate=crossover_rate_de,
        F=F_de
    )
    print(f"DE - Melhor: {best_de:.6f}, Pior: {worst_de:.6f}, Media: {avg_de:.6f}, Desvio Padrao: {std_de:.6f}")
    plot_convergence(mean_gen_de, std_gen_de, best_gen_de, worst_gen_de, "Convergencia DE - Min f(x) = x^2")
    de_optimal_hits['quadratic_min'] = count_optimal_hits(de_results_quad_min, 0.0, "min", tolerance)


    # --- Teste 2: Max f1(x) = x^2 ---
    print("\n--- Teste 2 (Algoritmo Genetico): Max f(x) = x^2 ---")
    best_ga, worst_ga, avg_ga, std_ga, mean_gen_ga, std_gen_ga, best_gen_ga, worst_gen_ga, ga_results_quad_max = run_experiment(
        algorithm_class=GeneticAlgorithm,
        objective_function=quadratic_function,
        problem_type="max",
        n_variables=n,
        lower_bound=-10,
        upper_bound=10,
        population_size=population_size,
        generations=generations,
        crossover_rate=crossover_rate_ga,
        mutation_rate=mutation_rate_ga
    )
    print(f"GA - Melhor: {best_ga:.6f}, Pior: {worst_ga:.6f}, Media: {avg_ga:.6f}, Desvio Padrao: {std_ga:.6f}")
    plot_convergence(mean_gen_ga, std_gen_ga, best_gen_ga, worst_gen_ga, "Convergencia GA - Max f(x) = x^2")
    ga_optimal_hits['quadratic_max'] = count_optimal_hits(ga_results_quad_max, 200.0, "max", tolerance) # Max of x^2 in [-10,10] is 100

    print("\n--- Teste 2 (Evolucao Diferencial): Max f(x) = x^2 ---")
    best_de, worst_de, avg_de, std_de, mean_gen_de, std_gen_de, best_gen_de, worst_gen_de, de_results_quad_max = run_experiment(
        algorithm_class=DifferentialEvolution,
        objective_function=quadratic_function,
        problem_type="max",
        n_variables=n,
        lower_bound=-10,
        upper_bound=10,
        population_size=population_size,
        generations=generations,
        crossover_rate=crossover_rate_de,
        F=F_de
    )
    print(f"DE - Melhor: {best_de:.6f}, Pior: {worst_de:.6f}, Media: {avg_de:.6f}, Desvio Padrao: {std_de:.6f}")
    plot_convergence(mean_gen_de, std_gen_de, best_gen_de, worst_gen_de, "Convergencia DE - Max f(x) = x^2")
    de_optimal_hits['quadratic_max'] = count_optimal_hits(de_results_quad_max, 200.0, "max", tolerance)


    # --- Teste 3: Min f2(x) = Rastrigin ---
    print("\n--- Teste 3 (Algoritmo Genetico): Min f(x) = Rastrigin ---")
    best_ga, worst_ga, avg_ga, std_ga, mean_gen_ga, std_gen_ga, best_gen_ga, worst_gen_ga, ga_results_rastrigin_min = run_experiment(
        algorithm_class=GeneticAlgorithm,
        objective_function=rastrigin_function,
        problem_type="min",
        n_variables=n,
        lower_bound=-5,
        upper_bound=5,
        population_size=population_size,
        generations=generations,
        crossover_rate=crossover_rate_ga,
        mutation_rate=mutation_rate_ga
    )
    print(f"GA - Melhor: {best_ga:.6f}, Pior: {worst_ga:.6f}, Media: {avg_ga:.6f}, Desvio Padrao: {std_ga:.6f}")
    plot_convergence(mean_gen_ga, std_gen_ga, best_gen_ga, worst_gen_ga, "Convergencia GA - Min f(x) = Rastrigin")
    ga_optimal_hits['rastrigin_min'] = count_optimal_hits(ga_results_rastrigin_min, 0.0, "min", tolerance)


    print("\n--- Teste 3 (Evolucao Diferencial): Min f(x) = Rastrigin ---")
    best_de, worst_de, avg_de, std_de, mean_gen_de, std_gen_de, best_gen_de, worst_gen_de, de_results_rastrigin_min = run_experiment(
        algorithm_class=DifferentialEvolution,
        objective_function=rastrigin_function,
        problem_type="min",
        n_variables=n,
        lower_bound=-5,
        upper_bound=5,
        population_size=population_size,
        generations=generations,
        crossover_rate=crossover_rate_de,
        F=F_de
    )
    print(f"DE - Melhor: {best_de:.6f}, Pior: {worst_de:.6f}, Media: {avg_de:.6f}, Desvio Padrao: {std_de:.6f}")
    plot_convergence(mean_gen_de, std_gen_de, best_gen_de, worst_gen_de, "Convergencia DE - Min f(x) = Rastrigin")
    de_optimal_hits['rastrigin_min'] = count_optimal_hits(de_results_rastrigin_min, 0.0, "min", tolerance)

    print("\n--- Resumo de acertos no otimo (tolerancia de 1e-3) ---")
    print(f"Algoritmo Genetico - Min x^2: {ga_optimal_hits['quadratic_min']} / 30")
    print(f"Algoritmo Genetico - Max x^2: {ga_optimal_hits['quadratic_max']} / 30")
    print(f"Algoritmo Genetico - Min Rastrigin: {ga_optimal_hits['rastrigin_min']} / 30")
    print(f"Evolucao Diferencial - Min x^2: {de_optimal_hits['quadratic_min']} / 30")
    print(f"Evolucao Diferencial - Max x^2: {de_optimal_hits['quadratic_max']} / 30")
    print(f"Evolucao Diferencial - Min Rastrigin: {de_optimal_hits['rastrigin_min']} / 30")
```

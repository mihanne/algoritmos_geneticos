```python

# -*- coding: utf-8 -*-
import numpy as np
import random
import math
import matplotlib.pyplot as plt

class Individual:
    def __init__(self, n_variables, lower_bound, upper_bound):
        self.n_variables = n_variables
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.chromosome = self._generate_chromosome()
        self.fitness = 0.0

    def _generate_chromosome(self):
        return [random.uniform(self.lower_bound, self.upper_bound) for _ in range(self.n_variables)]

class DifferentialEvolution:
    def __init__(self, D, NP, CR, F, generations, lower_bound, upper_bound, objective_function, problem_type):
        self.D = D
        self.NP = NP
        self.CR = CR
        self.F = F
        self.generations = generations
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.objective_function = objective_function
        self.problem_type = problem_type
        self.population = []
        self.best_individual_overall = None
        self.best_fitness_overall = float('-inf') if problem_type == "max" else float('inf')

    def _initialize_population(self):
        self.population = [Individual(self.D, self.lower_bound, self.upper_bound) for _ in range(self.NP)]
        self._evaluate_population()

    def _evaluate_population(self, individuals=None):
        if individuals is None:
            individuals = self.population
        for individual in individuals:
            individual.fitness = self.objective_function(individual.chromosome)

    def _mutate_and_crossover(self, target_idx):
        target = self.population[target_idx]
        indices = list(range(self.NP))
        indices.remove(target_idx)
        a, b, c = random.sample(indices, 3)
        a, b, c = self.population[a].chromosome, self.population[b].chromosome, self.population[c].chromosome
        donor = [a[j] + self.F * (b[j] - c[j]) for j in range(self.D)]

        trial = list(target.chromosome)
        j_rand = random.randint(0, self.D - 1)
        for j in range(self.D):
            if random.random() < self.CR or j == j_rand:
                trial[j] = donor[j]
            trial[j] = max(self.lower_bound, min(self.upper_bound, trial[j]))
        return trial

    def run(self, snapshot_interval=14):
        self._initialize_population()
        history = []
        snapshots = {}

        for gen in range(self.generations):
            new_population = []
            for i in range(self.NP):
                target = self.population[i]
                trial_chromosome = self._mutate_and_crossover(i)

                trial = Individual(self.D, self.lower_bound, self.upper_bound)
                trial.chromosome = trial_chromosome
                self._evaluate_population([trial])

                if (self.problem_type == "min" and trial.fitness < target.fitness) or                    (self.problem_type == "max" and trial.fitness > target.fitness):
                    new_population.append(trial)
                else:
                    new_population.append(target)

            self.population = new_population
            current_best = min(self.population, key=lambda x: x.fitness) if self.problem_type == "min" else max(self.population, key=lambda x: x.fitness)
            if self.best_individual_overall is None or                (self.problem_type == "min" and current_best.fitness < self.best_fitness_overall) or                (self.problem_type == "max" and current_best.fitness > self.best_fitness_overall):
                self.best_individual_overall = current_best
                self.best_fitness_overall = current_best.fitness

            history.append(self.best_fitness_overall)

            if gen % snapshot_interval == 0 or gen == self.generations - 1:
                snapshots[gen] = [Individual(self.D, self.lower_bound, self.upper_bound) for _ in range(self.NP)]
                for i in range(self.NP):
                    snapshots[gen][i].chromosome = list(self.population[i].chromosome)
                    snapshots[gen][i].fitness = self.population[i].fitness

        return self.best_individual_overall, history, snapshots

def quadratic_function(x):
    return sum(val**2 for val in x)

def rastrigin_function(x, A=10):
    return A * len(x) + sum(xi**2 - A * math.cos(2 * math.pi * xi) for xi in x)

def run_experiment(objective_function, problem_type, n_variables, lower_bound, upper_bound,
                   population_size=50, generations=100, CR=0.8, F=0.8):
    results = []
    histories = []
    snapshots_list = []

    for i in range(30):
        algo = DifferentialEvolution(n_variables, population_size, CR, F, generations,
                                     lower_bound, upper_bound, objective_function, problem_type)
        best_solution, history, snapshots = algo.run()
        results.append(best_solution.fitness)
        histories.append(history)
        snapshots_list.append(snapshots)
        print(f"Execucao {i+1:02d} - Melhor: {best_solution.fitness:.6f}")

    best = min(results) if problem_type == "min" else max(results)
    worst = max(results) if problem_type == "min" else min(results)
    avg = np.mean(results)
    std = np.std(results)
    histories = np.array(histories)

    return best, worst, avg, std, np.mean(histories, axis=0), np.std(histories, axis=0),            np.min(histories, axis=0) if problem_type == "min" else np.max(histories, axis=0),            np.max(histories, axis=0) if problem_type == "min" else np.min(histories, axis=0),            results, snapshots_list

def plot_convergence(mean_gen, std_gen, best_gen, worst_gen, title):
    gens = list(range(1, len(mean_gen) + 1))
    plt.figure(figsize=(10, 6))
    plt.plot(gens, mean_gen, label='MÃ©dia', color='blue')
    plt.plot(gens, std_gen, label='Desvio PadrÃ£o', color='gray', linestyle='--')
    plt.plot(gens, best_gen, label='Melhor', color='green')
    plt.plot(gens, worst_gen, label='Pior', color='red')
    plt.title(title)
    plt.xlabel('GeraÃ§Ã£o')
    plt.ylabel('Fitness')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_contour(populations_snapshots, func, title, lower_bound, upper_bound, step=0.1):
    x = np.arange(lower_bound, upper_bound + step, step)
    y = np.arange(lower_bound, upper_bound + step, step)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[func([i, j]) for i in x] for j in y])

    for gen, pop in populations_snapshots.items():
        plt.figure(figsize=(8, 6))
        plt.contourf(X, Y, Z, levels=50, cmap='viridis')

        # CompatÃ­vel com objetos do tipo Individual e Ant
        pop_x = [getattr(ind, "chromosome", getattr(ind, "position"))[0] for ind in pop]
        pop_y = [getattr(ind, "chromosome", getattr(ind, "position"))[1] for ind in pop]

        plt.scatter(pop_x, pop_y, c='red', label='PopulaÃ§Ã£o', s=20)
        plt.title(f"{title} - GeraÃ§Ã£o {gen}")
        plt.xlabel("x1")
        plt.ylabel("x2")
        plt.colorbar(label="Fitness")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

def count_optimal_hits(results, optimal, problem_type, tolerance=1e-3):
    return sum(abs(r - optimal) <= tolerance for r in results)

class Ant:
    def __init__(self, position):
        self.position = position
        self.fitness = None

class AntColonyOptimization:
    def __init__(self, n_ants, generations, alpha, beta, rho, n_variables,
                 lower_bound, upper_bound, objective_function, problem_type):
        self.n_ants = n_ants
        self.generations = generations
        self.alpha = alpha
        self.beta = beta
        self.rho = rho
        self.n_variables = n_variables
        self.lower_bound = lower_bound
        self.upper_bound = upper_bound
        self.objective_function = objective_function
        self.problem_type = problem_type
        self.archive = []

    def _weighted_sample(self):
        fitnesses = np.array([a.fitness for a in self.archive])
        if self.problem_type == "max":
            weights = np.exp(self.alpha * (fitnesses - np.max(fitnesses)))
        else:
            weights = np.exp(-self.alpha * (fitnesses - np.min(fitnesses)))
        weights /= weights.sum()
        return random.choices(self.archive, weights=weights, k=1)[0]

    def _generate_solution(self):
        if not self.archive:
            return [random.uniform(self.lower_bound, self.upper_bound) for _ in range(self.n_variables)]

        elite = self._weighted_sample()
        stds = np.std([a.position for a in self.archive], axis=0)
        stds = [s if s > 1e-5 else 0.1 for s in stds]
        position = [random.gauss(mu, s / self.beta) for mu, s in zip(elite.position, stds)]
        return [min(max(x, self.lower_bound), self.upper_bound) for x in position]

    def run(self, snapshot_interval=14):
        best_ant = None
        best_fitness = float("-inf") if self.problem_type == "max" else float("inf")
        history = []
        snapshots = {}

        for gen in range(self.generations):
            ants = []
            for _ in range(self.n_ants):
                pos = self._generate_solution()
                ant = Ant(pos)
                ant.fitness = self.objective_function(ant.position)
                ants.append(ant)

            if self.problem_type == "min":
                ants.sort(key=lambda x: x.fitness)
            else:
                ants.sort(key=lambda x: -x.fitness)

            self.archive = ants[:int(self.n_ants * (1 - self.rho))]

            if (self.problem_type == "min" and ants[0].fitness < best_fitness) or                (self.problem_type == "max" and ants[0].fitness > best_fitness):
                best_ant = ants[0]
                best_fitness = ants[0].fitness

            history.append(best_fitness)
            if gen % snapshot_interval == 0 or gen == self.generations - 1:
                snapshots[gen] = ants

        return best_ant, history, snapshots

def rastrigin_function(x, A=10):
    return A * len(x) + sum(xi**2 - A * math.cos(2 * math.pi * xi) for xi in x)

def quadratic_function(x):
    return sum(val**2 for val in x)

def run_experiment_aco(objective_function, problem_type, n_variables, lower_bound, upper_bound,
                       n_ants=50, generations=100, alpha=1.0, beta=1.0, rho=0.1):
    results = []
    histories = []
    snapshots_list = []

    for i in range(30):
        aco = AntColonyOptimization(n_ants, generations, alpha, beta, rho,
                                    n_variables, lower_bound, upper_bound, objective_function, problem_type)
        best_solution, history, snapshots = aco.run()
        results.append(best_solution.fitness)
        histories.append(history)
        snapshots_list.append(snapshots)
        print(f"Execucao {i+1:02d} - Melhor: {best_solution.fitness:.6f}")

    best = min(results) if problem_type == "min" else max(results)
    worst = max(results) if problem_type == "min" else min(results)
    avg = np.mean(results)
    std = np.std(results)
    histories = np.array(histories)

    return best, worst, avg, std, np.mean(histories, axis=0), np.std(histories, axis=0), \
           np.min(histories, axis=0) if problem_type == "min" else np.max(histories, axis=0), \
           np.max(histories, axis=0) if problem_type == "min" else np.min(histories, axis=0), \
           results, snapshots_list

if __name__ == "__main__":
    n = 2
    pop = 50
    gens = 100
    CR = 0.8
    F = 0.8

    print("\n--- Teste 1: Min f(x) = x^2 ---")
    best, worst, avg, std, mean_gen, std_gen, best_gen, worst_gen, results, snapshots_list = run_experiment(
        quadratic_function, "min", n, -10, 10, pop, gens, CR, F)
    print(f"Melhor: {best:.6f}, Pior: {worst:.6f}, MÃ©dia: {avg:.6f}, Desvio: {std:.6f}")
    plot_convergence(mean_gen, std_gen, best_gen, worst_gen, "DE - Min f(x)=xÂ²")
    #plot_contour(snapshots_list[0], quadratic_function, "Curvas - Min xÂ²", -10, 10)
    print(f"Acertos Ã³timos (<=1e-3): {count_optimal_hits(results, 0.0, 'min')} / 30")

    print("\n--- Teste 2: Max f(x) = x^2 ---")
    best, worst, avg, std, mean_gen, std_gen, best_gen, worst_gen, results, snapshots_list = run_experiment(
        quadratic_function, "max", n, -10, 10, pop, gens, CR, F)
    print(f"Melhor: {best:.6f}, Pior: {worst:.6f}, MÃ©dia: {avg:.6f}, Desvio: {std:.6f}")
    plot_convergence(mean_gen, std_gen, best_gen, worst_gen, "DE - Max f(x)=xÂ²")
    #plot_contour(snapshots_list[0], quadratic_function, "Curvas - Max xÂ²", -10, 10)
    print(f"Acertos Ã³timos (>=99.999): {count_optimal_hits(results, 100.0, 'max')} / 30")

    print("\n--- Teste 3: Min f(x) = Rastrigin ---")
    best, worst, avg, std, mean_gen, std_gen, best_gen, worst_gen, results, snapshots_list = run_experiment(
        rastrigin_function, "min", n, -5, 5, pop, gens, CR, F)
    print(f"Melhor: {best:.6f}, Pior: {worst:.6f}, MÃ©dia: {avg:.6f}, Desvio: {std:.6f}")
    plot_convergence(mean_gen, std_gen, best_gen, worst_gen, "DE - Min Rastrigin")
    #plot_contour(snapshots_list[0], rastrigin_function, "Curvas - Min Rastrigin", -5, 5)
    print(f"Acertos Ã³timos (<=1e-3): {count_optimal_hits(results, 0.0, 'min')} / 30")


    print("\n=== ACO - Teste 1: Min f(x) = x^2 ===")
    best, worst, avg, std, mean_gen, std_gen, best_gen, worst_gen, results, snapshots_list = run_experiment_aco(
        quadratic_function, "min", n, -10, 10)
    print(f"Melhor: {best:.6f}, Pior: {worst:.6f}, MÃ©dia: {avg:.6f}, Desvio: {std:.6f}")
    plot_convergence(mean_gen, std_gen, best_gen, worst_gen, "ACO - Min f(x)=xÂ²")
    #plot_contour(snapshots_list[0], quadratic_function, "ACO - Curvas - Min xÂ²", -10, 10)
    print(f"Acertos Ã³timos (<=1e-3): {count_optimal_hits(results, 0.0, 'min')} / 30")

    print("\n=== ACO - Teste 2: Max f(x) = x^2 ===")
    best, worst, avg, std, mean_gen, std_gen, best_gen, worst_gen, results, snapshots_list = run_experiment_aco(
        quadratic_function, "max", n, -10, 10)
    print(f"Melhor: {best:.6f}, Pior: {worst:.6f}, MÃ©dia: {avg:.6f}, Desvio: {std:.6f}")
    plot_convergence(mean_gen, std_gen, best_gen, worst_gen, "ACO - Max f(x)=xÂ²")
    #plot_contour(snapshots_list[0], quadratic_function, "ACO - Curvas - Max xÂ²", -10, 10)
    print(f"Acertos Ã³timos (>=99.999): {count_optimal_hits(results, 100.0, 'max')} / 30")

    print("\n=== ACO - Teste 3: Min f(x) = Rastrigin ===")
    best, worst, avg, std, mean_gen, std_gen, best_gen, worst_gen, results, snapshots_list = run_experiment_aco(
        rastrigin_function, "min", n, -5, 5)
    print(f"Melhor: {best:.6f}, Pior: {worst:.6f}, MÃ©dia: {avg:.6f}, Desvio: {std:.6f}")
    plot_convergence(mean_gen, std_gen, best_gen, worst_gen, "ACO - Min Rastrigin")
    #plot_contour(snapshots_list[0], rastrigin_function, "ACO - Curvas - Min Rastrigin", -5, 5)
    print(f"Acertos Ã³timos (<=1e-3): {count_optimal_hits(results, 0.0, 'min')} / 30")

```
